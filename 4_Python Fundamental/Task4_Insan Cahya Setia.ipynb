## 1. Buat Kalkulator Luas

```python
def Kalkulator_luas(luas, panjang, lebar):
    luas = luas.lower()
    
    if luas == 'bujur sangkar':
        luas_BujurSangkar = panjang * lebar
        print('Luas {} adalah {} cm^2'.format(luas, luas_BujurSangkar))
    
    elif luas == 'segitiga':
        luas_Segitiga = 0.5 * lebar * panjang
        print('Luas {} adalah {} cm^2'.format(luas, luas_Segitiga))
    
    elif luas == 'jajar genjang':
        luas_JajarGenjang = lebar * panjang
        print('Luas {} adalah {} cm^2'.format(luas, luas_JajarGenjang))
    
    else:
        print('Masukkan luas dengan benar')
```

```python
def Luas_trapesium(a, b, t):
        luas_Trapesium = 0.5 * (a + b) * t
        print('Luas Trapesium adalah {} cm^2'.format(luas_Trapesium))
```

```python
Kalkulator_luas('Bujur Sangkar', 4, 3)
```

```python
Luas_trapesium(3, 5, 4)
```

## 2. Buatlah 1 function tanpa parameter dan terdapat 1 variable, dimana kamu melakukan logaritma pada variabel tersebut

```python
def logaritma():
    angka = int(input('Masukkan angka '))
    hasil = mt.log(angka)
    return 'Logaritma dari {} adalah {}'.format(angka, hasil)
```

## Data Banking

```python
banking = pd.read_csv('Banking.csv')
banking.head()
```

```python
banking.info()
```

```python
# Create list
col_status = []

# Define looping for status column
for x, col in banking.iterrows():
    if col['job'] == 'management':
        if col['education'] == 'university.degree':
            if col['duration'] < 150:
                status = 'rich'
            else:
                status = 'unknown'
        else:
            status = 'unknown'
            
    elif col['job'] == 'technician':
        if col['education'] == 'university.degree':
            if col['duration'] >= 150 & col['duration'] <= 200:
                status = 'middle'
            else:
                status = 'unknown'
        else:
            status = 'unknown'
    
    elif col['job'] == 'blue-collar':
        if col['education'] == 'basic.4y' or col['education'] == 'basic.6y' or col['education'] == 'basic.9y':
            if col['duration'] >200:
                status = 'middle'
            else:
                status = 'unknown'
        else:
            status = 'unknown'
    else:
        status = 'unknown'
    
    col_status.append(status)
    
    
banking['status'] = col_status
```
